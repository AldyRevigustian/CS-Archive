```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
x = [1,2,3,4,5,6,7,8]
y = [1,5,7,9,11,13,16, 20]
# x = [1,2,3,4,5,6,7,8]
# y = [2,5,1,7,9,3,6, 10]

x = np.array(x)
y = np.array(y)


A = np.vstack([x, np.ones(len(x))]).T
# y = mx + b
#     ^ = A
# print(A)
y = y[:, np.newaxis]

# Direct inverse 
beta = np.dot(np.dot(np.linalg.inv(np.dot(A.T, A)), A.T), y)
plt.title("Direct inverse")
plt.plot(x,y, 'bo')
plt.plot(x, beta[0]*x + beta[1], 'r') # y = mx + b
plt.show()

# Pseudo Inverse
pinv = np.linalg.pinv(A)
beta2 = pinv.dot(y)
print(pinv)
plt.title("Pesudo Inverse")
plt.plot(x,y, 'o')
plt.plot(x, beta2[0]*x + beta2[1], 'r') # y = mx + b
plt.show()
```
