```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,3,5,6,8]
x = np.array(x)
y = np.array(y)

A = np.vstack([x, np.ones(len(x))]).T
y = y[:, np.newaxis]

pinv = np.linalg.pinv(A)
beta = pinv.dot(y)

plt.title("Title")
plt.plot(x,y, 'bo')
plt.plot(x, beta[0] * x + beta[1], 'r')
plt.show()
```
